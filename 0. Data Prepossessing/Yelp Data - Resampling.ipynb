```python
# seed random number generator
seed(1)
# prepare a sequence
sequence = [i for i in range(len(non_target_business_id))]

# select a subset without replacement
subset = sample(sequence, 16312)
```

```python
non_target_id_list = []
for item in subset:
    this_business_id = non_target_business_id[item]
    non_target_id_list.append(this_business_id)
```
